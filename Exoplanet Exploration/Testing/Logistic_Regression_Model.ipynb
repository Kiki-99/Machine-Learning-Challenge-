```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (Columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[["koi_period", "koi_period_err1", "koi_period_err2", "koi_time0bk", "koi_time0bk_err1"]]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
y = df["koi_disposition"]
X = df.drop(columns=["koi_disposition"])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

# Pre-Processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Create the Model



```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression Model
model = LogisticRegression(max_iter=1000)

# Fit (train) or model using the training data
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
# print prediction results
from sklearn.metrics import classification_report

predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions))
```

## Indentifying Insignificant Variables 

```python
# from sklearn.svm import SVC
from sklearn.feature_selection import RFE

# Create the RFE model
rfe = RFE(model)

# Fit model
rfe = rfe.fit(X_train_scaled, y_train)

# summarize the selection of the attributes
rfe_df = pd.DataFrame({
    'features': X.columns,
    'support': rfe.support_,
    'ranking': rfe.ranking_
})
rfe_df
```

```python
# Identify insignificant features
features = rfe_df.loc[rfe_df['support'] == False]
drop_columns = list(features['features'])
```

## Training the Model

```python
#Using only significant features
drop_columns.append('koi_disposition')
drop_columns
```

```python
# Set features as X value and label as y value
X = df.drop(columns=drop_columns)
y = df['koi_disposition']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scale your data
X_minmax = MinMaxScaler().fit(X_train)

X_train_scaled = X_minmax.transform(X_train)
X_test_scaled = X_minmax.transform(X_test)
```

```python
# Create a Logistic Regression Model
model = LogisticRegression(max_iter=3000)

# Fit (train) or model using the training data
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
# print prediction results
from sklearn.metrics import classification_report

predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
from sklearn.exceptions import FitFailedWarning
import warnings
warnings.simplefilter('always', FitFailedWarning)
param_grid = {'C': [1, 5, 10],
              'penalty': ["l1", "l2"]}

grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
with warnings.catch_warnings(record=True) as w:
    try:
        grid.fit(X_train_scaled, y_train) 
    except ValueError:
         pass
```

```python
print(f'Best Parameter: {grid.best_params_}')
print(f'Best Model Score: {grid.best_score_}')
```

```python
# print prediction results
from sklearn.metrics import classification_report

predictions = grid.predict(X_test_scaled)
print(classification_report(y_test, predictions))
```
